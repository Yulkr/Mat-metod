```python
def l2_opim_prib_monopolista():
    a,b,c,d,p1,p2,l,m,n,k = symbols('a b c d p1 p2 l m n k ') 
    a1,b1,c1,d1,l1,m1,n1,k1 = symbols('a1 b1 c1 d1 l1 m1 n1 k1 ') 
    q1 = eval(input("Введите функцию спроса q1, например  a+b*p1:"))
    q2 = eval(input("Введите функцию спроса q2, например  a+b*p2:"))
    C = eval(input("Введите функцию издержек С, например k*(q1)**2+l*q1*q2+m*(q2)**2+n:"))
    a1 = int(input("a"))
    b1 = int(input("b"))
    c1 = int(input("c"))
    d1 = int(input("d"))
    k1 = int(input("k"))
    l1 = int(input("l"))
    m1 = int(input("m"))
    n1 = int(input("n"))
    
    print("Вывод С:", C)
    P = p1*q1+p2*q2-C
    P1 = P.subs([(a,a1),(b,b1),(c,c1),(d,d1),(k,k1),(l,l1),(m,m1),(n,n1)])
    print("Вывод P:",P1)
    #Находим производную
    proizvid1 = P1.diff(p1)
    print('df/dp1 =', proizvid1)
    proizvid2 =  P1.diff(p2)
    print('df/dp2 =', proizvid2)
    sols = solve([proizvid1,proizvid2], p1,p2,dict=True)
    print(' Точка M(p1,p2) -', sols)
    print("p1", sols[0][p1])
    p11 = sols[0][p1]
    print("p2", sols[0][p2])
    p22 =sols[0][p2]
    p111 = int(p11)
    p222 = int(p22)
    #Подставляем значение
    q1 = q1.subs([(a,a1),(b,b1),(c,c1),(d,d1),(k,k1),(l,l1),(m,m1),(n,n1),(p1,p11),(p2,p22)])
    q2= q2.subs([(a,a1),(b,b1),(c,c1),(d,d1),(k,k1),(l,l1),(m,m1),(n,n1),(p1,p11),(p2,p22)])
    C = C.subs([(a,a1),(b,b1),(c,c1),(d,d1),(k,k1),(l,l1),(m,m1),(n,n1),(p1,p11),(p2,p22)])
    P2 = P.subs([(a,a1),(b,b1),(c,c1),(d,d1),(k,k1),(l,l1),(m,m1),(n,n1),(p1,p11),(p2,p22)])
    print("Значение q1 при максимальном p1:", q1,"Значение q2 при максимальном p2:", q2)
    print("Значение С при максимальном p1 и p2:",C)
    print("значение P при максимальном p1 и p2:",P2)
    #3D график
    plt.figure()
    ax = plt.axes(projection='3d')
    x = np.linspace(-p111*0.2,p111*2, p111*10)
    y = np.linspace(-p222*0.2,p222*2,p222*10)
    X,Y = np.meshgrid(x,y)
    Z = lambdify(["p1","p2"], P1)(X,Y)
    ax.plot_surface(X,Y,Z,cmap='viridis')
    ax.scatter(p11,p22, P2, s =10, color="red", label='М*({p11},{p22})'.format(p11=p11, p22=p22))
    plt.title('Прибыль фирмы монополиста')
    plt.xlabel=("Цена единицы товара x1")
    plt.ylabel=("Цена единицы товара x2")
    plt.legend()
    plt.show()

    #График №2
    x,y = np.mgrid[-0.5*p11:2*p11,-0.5*p22:2*p22]                
    z = lambdify(["p1","p2"], P1)(x,y)
    fig, ax = plt.subplots()
    ax.scatter(p11,p22, color='red')
    plt.text(p11,p22,'М*({p11},{p22})'.format(p11=p11, p22=p22))
    ax.contour(x,y,z, levels = 30)
    fig.set_figwidth(6)    
    fig.set_figheight(6)   

    plt.show()  

l2_opim_prib_monopolista() 
```
